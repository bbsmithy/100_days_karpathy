```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 -4*x + 5
```

```python
f(3)
```

This is a quadratic function. To plot this function we need a set of x values and a setof y values with each Y value being the corresponding f(x) value.

We can calculate the y value for each corresponding x value by using the function we just defined. So we'll have an array of x values and y values each index corresponding to each other.

```python
xs = np.arange(-5.0, 5.0, 0.25)
xs
```

```python
ys = f(xs)
ys
```

```python
plt.plot(xs, ys)
plt.show()
```

Lets get the derivative of the function. What is the derivative? It's the slope of the function at any given point. The derivative of a function f(x) at a point x is the instantaneous rate of change of the function at that point or (the slope of the tangent line to the function at that point).

```python
h = 0.001 # increase in x (run)
x = 3.0 # point of interest
print("f(x) = ", f(x)) # the y value at x
print("f(x+h) = ", f(x+h)) # the y value at x+h
```

We can see that when we add a small poitsive value to x, the function value (Y value on the graph) increases by roughly .014. In other words when we increase the x value by a small amount the y value also increases by a small amount.

```python
print("f(x+h) - f(x) = ", f(x+h) - f(x), " which is the change in Y when X is increased by ", h)
print("(f(x+h) - f(x))/h = ", (f(x+h) - f(x))/h, " which is the slope of the function at point ", x)
```

We get the difference between f(x+h) and f(x) to get the change in value Y when H is added to X in the function call. We divide this by H to get the slope of the function at that point. The slope of the function being the rise (Increase in Y value: 0.014) over the run (the change along the X axis: 0.001). Lets change X to a minus value and see what happens.

```python
h = 0.001 # increase in x (run)
x = -3.0 # point of interest
print("X value: ", x)
print("H value: ", h)
print("f(x) = ", f(x), "the y value at x") # the y value at x
print("f(x+h) = ", f(x+h), "the y value at x+h") # the y value at x+h
print("f(x+h) - f(x) = ", f(x+h) - f(x), " which is the change in Y when X is increased by ", h)
print("(f(x+h) - f(x))/h = ", (f(x+h) - f(x))/h, " which is the slope of the function at point ", x)

# display the slop at -3
slope_at_x = (f(x+h) - f(x))/h
print("Slope at x: ", slope_at_x)
plt.plot(xs, ys)
plt.show()
```

The slope is -21.996999999998934 when x is -3.0 and h is 0.00, for the function f(x) = 3x - 4x + 5. Lets visualise the slope or tangent line when x is -3.0 and h is 0.001.

```python
# ... existing code ...
x = -3.0
h = 0.001
slope_at_x = (f(x+h) - f(x))/h

plt.figure(figsize=(10, 6))
plt.plot(xs, ys, label='f(x) = 3x - 4x + 5')

# Calculate points for the tangent line
tangent_xs = np.array([x - 2, x + 2])
tangent_ys = slope_at_x * (tangent_xs - x) + f(x)

plt.plot(tangent_xs, tangent_ys, 'r--', label=f'Tangent line at x = {x}')
plt.plot(x, f(x), 'ro', label='Point of tangency')

plt.title(f'Function and its Tangent Line at x = {x}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Print the slope for reference
print(f"Slope of the tangent line at x = {x}: {slope_at_x}")
```
